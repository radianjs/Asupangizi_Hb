```python
import pandas as pd
import numpy as np
import sklearn.model_selection as ms
import matplotlib.pyplot as plt
df1 = pd.read_csv('E:\Belajar Python\Data set\dataHB.csv')
AKG=df1[['AKG']]
HB=df1[['HB']]
X_train, X_test, y_train, y_test = ms.train_test_split(AKG, HB, test_size=0.2, random_state=0)
plt.scatter(X_train, y_train, edgecolors='r')
plt.xlabel('AKG')
plt.ylabel('HB')
plt.title('Hubungan AKG dengan HB')
x1 = np.linspace(60,95)
y1 = 0.0045 + 0.0985* x1
plt.plot(x1,y1)
plt.show()

import sklearn.linear_model as lm
model1 = lm.LinearRegression()
model1.fit(X_train, y_train)
print(model1.coef_)

print(model1.intercept_)

plt.scatter(X_train, y_train, edgecolors='r')
plt.xlabel('AKG')
plt.ylabel('HB')
plt.title('Hubungan AKG dengan HB')
x1 = np.linspace(60,95)
y1 = -4.37144041 + 0.1617675* x1
plt.plot(x1,y1)
plt.show()

r2 = model1.score(X_test, y_test)
print(r2)

prediksi = model1.predict(X_test)
print(prediksi)
print(y_test)

np.concatenate((prediksi, y_test),axis=1)
```
